# Vipassana Course Scraper

## 1. Import Dependencies

```python
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

## 2. Navigate to Site and Expand Location List

```python
url = "https://www.dhamma.org/en-US/locations/directory#directory-list"
browser.visit(url)
browser.click_link_by_text('Expand All')
```

```python
soup = BeautifulSoup(browser.html, 'lxml')
print(soup.prettify())
```

## 3. Extract Info for Each Location

```python
directory = soup.find('div', class_="regions")
locations = directory.find_all('div', class_="location")
for location in locations:
    location_name = location.find('span', class_="display-name")
    if location.find('a', class_="location-courses-link")['href'] != "#":
        location_link = location.find('a', class_="location-courses-link")
    else:
        dropdown = location.find('ul', class_="dropdown-menu")
        links = dropdown.find_all('a')
        location_link = [link for link in links if "English" in link.text][0]
    print(location_name.text, location_link['href'])
```
